# Wine Dataset Analysis

~ Write a Python program to load the wine dataset using the Sklearn library to a DataFrame and perform the following tasks:

1.Convert the dataset into DataFrame using pandas.

2.Generate the sample size of 50 and give a random state as 100.

3.Calculate Z-critical, Margin of Error, and Confidence Interval for alcohol at 95% significance interval on generated sample data.

```python
import pandas as pd
from sklearn.datasets import load_wine

# Step 1: Load the wine dataset
wine_data = load_wine()

# Step 2: Convert the dataset into a DataFrame
wine_df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

# Step 3: Generate a sample of size 50 with a random state of 100
sample_size = 50
random_state = 100
sample_df = wine_df.sample(n=sample_size, random_state=random_state)

# Display the sample DataFrame
print("Sample DataFrame:")
print(sample_df)
```

```python
import numpy as np
import scipy.stats as stats

# Assuming you have calculated the mean and standard deviation on the sample
sample_mean = sample_df['alcohol'].mean()  # replace 'alcohol' with the appropriate column name
sample_std = sample_df['alcohol'].std()
n = sample_df.shape[0]

# Calculate Z-critical value for 95% confidence (two-tailed)
z_critical = stats.norm.ppf(0.975)

# Calculate Margin of Error
margin_of_error = z_critical * (sample_std / np.sqrt(n))

# Calculate Confidence Interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Z-critical value: {z_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: {confidence_interval}")
```
